```python
# === Imports ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Matplotlib- und Seaborn-Einstellungen
plt.rcParams['figure.figsize'] = (10, 6)
sns.set(style="whitegrid")

# === Funktion zur Speicherung der Plots ===
def save_plot(filename, file_format="png"):
    """
    Speichert den aktuellen Plot als Datei.
    :param filename: Der Name der Datei (ohne Erweiterung)
    :param file_format: Das Dateiformat, z.B. 'png' oder 'jpeg' (Standard: 'png')
    """
    plt.savefig(f"{filename}.{file_format}", format=file_format, dpi=300)
    print(f"Plot gespeichert als {filename}.{file_format}")

# === Funktionen zur Exploration numerischer Spalten ===
def explore_numerical_column(df, col):
    """
    Zeigt deskriptive Statistiken, ein Histogramm (ohne KDE) und einen Boxplot mit
    eingezeichnetem Mittelwert (rot) und Median (blau) für die Spalte.
    """
    print(f"--- Analyse numerische Spalte: {col} ---")
    print(df[col].describe())
    
    mean_val = df[col].mean()
    median_val = df[col].median()
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Histogramm ohne KDE
    sns.histplot(df[col], kde=False, ax=axes[0], color="skyblue")
    axes[0].axvline(mean_val, color="red", linestyle="--", label=f"Mittelwert: {mean_val:.2f}")
    axes[0].axvline(median_val, color="blue", linestyle="--", label=f"Median: {median_val:.2f}")
    axes[0].set_title(f"Histogramm von {col}")
    axes[0].legend()
    
    # Boxplot
    sns.boxplot(x=df[col], ax=axes[1], color="lightgreen")
    axes[1].axvline(mean_val, color="red", linestyle="--")
    axes[1].axvline(median_val, color="blue", linestyle="--")
    axes[1].set_title(f"Boxplot von {col}")
    
    plt.tight_layout()

    # Speichern der Plots
    save_plot(f"{col}_plot", file_format="png") 
    
    plt.show()

# === Funktion zur Exploration kategorialer Spalten ===
def explore_categorical_column(df, col):
    """
    Zeigt die Häufigkeiten und erstellt ein Balkendiagramm (Countplot) für die Spalte.
    """
    print(f"--- Analyse kategoriale Spalte: {col} ---")
    print(df[col].value_counts(dropna=False))
    
    plt.figure(figsize=(8, 5))
    order = df[col].value_counts(dropna=False).index
    sns.countplot(data=df, x=col, order=order, palette="viridis")
    plt.title(f"Anzahl der Beobachtungen pro Kategorie in {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()

    save_plot(f"{col}_countplot", file_format="png")
    
    plt.show()

# === Korrelationsmatrix-Plot ===
def plot_correlation_matrix(df, numeric_cols, target_col):
    """
    Erstellt eine Heatmap der Korrelationsmatrix der angegebenen numerischen Spalten (inkl. Zielspalte).
    """
    cols = numeric_cols.copy()
    if target_col not in cols:
        cols.append(target_col)
    
    corr_matrix = df[cols].corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
    plt.title("Korrelationsmatrix (inkl. Zielvariable)")
    plt.tight_layout()

    save_plot(f"{col}_countplot", file_format="png")
    
    plt.show()

# === Pairplot (Multiplot) zur Visualisierung der Korrelationen ===
def plot_pairplot(df, numeric_cols, target_col, exclude_columns=[]):
    """
    Erstellt ein Pairplot (Streudiagramm-Matrix) für die angegebenen numerischen Spalten 
    (ohne ausgeschlossene Spalten) inklusive Zielspalte.
    """
    # Kopie, um nichts am Original zu verändern
    cols = [col for col in numeric_cols if col not in exclude_columns]
    if target_col not in cols:
        cols.append(target_col)
    
    # Pairplot
    sns.pairplot(df[cols], diag_kind="hist")
    plt.suptitle("Pairplot der numerischen Spalten (inkl. Zielvariable)", y=1.02)

    save_plot(f"{col}_countplot", file_format="png")
    
    plt.show()

# === Scatter- und Boxplot-Visualisierungen zur Beziehung mit der Zielspalte ===
def plot_target_relationships(df, target_col, exclude_columns=[]):
    """
    Erstellt Scatter-Plots für numerische Features (außer Zielspalte und den ausgeschlossenen Spalten)
    und Boxplots für kategoriale Features, jeweils gegen die Zielspalte.
    
    Zusätzlich wird der Pearson-Korrelationskoeffizient für numerische Features ausgegeben.
    """
    # Bestimme numerische und kategoriale Spalten
    numeric_cols = list(df.select_dtypes(include=[np.number]).columns)
    categorical_cols = list(df.select_dtypes(include=['object']).columns)
    
    # Entferne die Zielspalte, falls sie numerisch ist
    if target_col in numeric_cols:
        numeric_cols.remove(target_col)
    
    # Ausschließen der gewünschten Spalten (z. B. Blood Pressure)
    numeric_cols = [col for col in numeric_cols if col not in exclude_columns]
    categorical_cols = [col for col in categorical_cols if col not in exclude_columns and col != target_col]
    
    print(f"\n=== Scatter-Plots (numerische Features vs {target_col}) ===")
    for col in numeric_cols:
        corr = df[col].corr(df[target_col])
        print(f"Korrelation zwischen {col} und {target_col}: {corr:.2f}")
        
        plt.figure(figsize=(8, 6))
        sns.regplot(x=df[col], y=df[target_col], scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
        plt.title(f"{col} vs {target_col} (r = {corr:.2f})")
        plt.tight_layout()

        save_plot(f"{col}_countplot", file_format="png")
        
        plt.show()
    
    print(f"\n=== Boxplots (kategoriale Features vs {target_col}) ===")
    for col in categorical_cols:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[col], y=df[target_col], palette="pastel")
        plt.title(f"{col} vs {target_col}")
        plt.xticks(rotation=45)
        plt.tight_layout()

        save_plot(f"{col}_countplot", file_format="png")
        
        plt.show()

# === Gesamte Datenexploration ===
def run_exploration(df):
    """
    Zeigt grundlegende Informationen und führt die Exploration der einzelnen Spalten durch.
    Gibt ein modifiziertes DataFrame (df_expl) und Listen numerischer und kategorialer Spalten zurück.
    """
    print("=== Grundlegende Dateninformationen ===")
    print(df.info())
    
    # Falls vorhanden, Patient ID entfernen
    df_expl = df.copy()
    if 'Patient ID' in df_expl.columns:
        df_expl = df_expl.drop(columns=['Patient ID'])
    
    # Spalten anhand des Datentyps aufteilen
    categorical_cols = list(df_expl.select_dtypes(include=['object']).columns)
    numerical_cols = list(df_expl.select_dtypes(include=[np.number]).columns)
    
    # Numerische Spalten mit genau 2 eindeutigen Werten als kategorial behandeln
    for col in numerical_cols.copy():
        if df_expl[col].nunique() == 2:
            categorical_cols.append(col)
            numerical_cols.remove(col)
    
    print("\nGefundene numerische Spalten:", numerical_cols)
    print("Gefundene kategoriale Spalten:", categorical_cols)
    
    # Exploration numerischer Spalten
    for col in numerical_cols:
        explore_numerical_column(df_expl, col)
    
    # Exploration kategorialer Spalten
    for col in categorical_cols:
        explore_categorical_column(df_expl, col)
    
    return df_expl, numerical_cols, categorical_cols

# === Hauptfunktion ===
def main():
    # === Datensatz laden (Pfad ggf. anpassen) ===
    csv_path = "data.csv"
    df = pd.read_csv(csv_path)
    print(f"Datensatz erfolgreich geladen. Form: {df.shape}")
    
    # Gesamte Exploration durchführen
    df_expl, numerical_cols, categorical_cols = run_exploration(df)
    
    # Zielspalte definieren (bitte anpassen, falls Name anders)
    target_col = "Heart Attack Risk"
    
    # Spalte "Blood Pressure" wird ausgeschlossen, da sie sehr viele Ausprägungen hat
    exclude_for_scatter = ["Blood Pressure"]
    
    # Korrelationsmatrix der numerischen Spalten (ohne 'Blood Pressure') inkl. Zielspalte
    numeric_for_corr = [col for col in numerical_cols if col not in exclude_for_scatter]
    if target_col not in numeric_for_corr:
        numeric_for_corr.append(target_col)
    
    # 1) Heatmap der Korrelationen
    plot_correlation_matrix(df_expl, numeric_for_corr, target_col)
    
    # 2) Pairplot (Multiplot) der numerischen Spalten inkl. Zielspalte
    plot_pairplot(df_expl, numeric_for_corr, target_col, exclude_columns=[])
    
    # Scatter- und Boxplots zu den Beziehungen der Features mit der Zielspalte
    plot_target_relationships(df_expl, target_col, exclude_columns=exclude_for_scatter)
    
    print("\n=== Datenexploration abgeschlossen ===")

if __name__ == "__main__":
    main()
```
